# 1. Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans are:", valid_plans)
        return False
    else:
        print("Recharge successful for ₹", amount)
        return True


# Main Program
while True:
    try:
        amount = int(input("Enter recharge amount: "))
        
        if validate_recharge(amount):
            break   # Exit loop if valid
        
        print("Please try again.\n")
    
    except ValueError:
        print("Please enter a valid number.\n")
```

# 2. Inventory Reorder Alert System

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"Reorder Alert: {product} has only {quantity} items left.")
        else:
            print(f"Stock OK: {product} has {quantity} items.")

inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18,
    "USB Cable": 5
}

check_inventory(inventory)
```

# 3. Student Result Processing System

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark
    return total / len(marks)


marks = []
n = int(input("Enter number of subjects: "))

for i in range(n):
    m = int(input(f"Enter mark {i+1}: "))
    marks.append(m)

avg = process_result(marks)

print("Average Marks:", avg)

if avg >= 50:
    print("Result: Pass")
else:
    print("Result: Fail")
```

# 4. Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)   # Base + per km

    if peak == "yes":
        fare = fare + (fare * 0.25)   # Add 25% extra

    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total = calculate_fare(distance, peak)

    print("Total Fare = ₹", total)

    choice = input("Do you want to try again? (yes/no): ")

    if choice == "no":
        print("Thank you!")
        break
```

# 5. Employee Attendance Eligibility

```python
def check_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

result = check_eligibility(attendance)

print("Attendance Result:", result)
```

# 6. Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter your password: ")
result = check_password(pwd)

print(result)
```
